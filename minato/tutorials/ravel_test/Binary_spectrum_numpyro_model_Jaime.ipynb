```python
def sb2_model(wavelength=None, fluxes=None):

    n_epoch, n_wavelength = fluxes.shape

    #spectral window
    cen = numpyro.param("cen", (wavelength[0].max() + wavelength[0].min())/2)
    delta_cen = numpyro.param("delta_cen", (wavelength[0].max() - wavelength[0].min()))

    # Model parameters
    continuum = numpyro.sample('continuum', dist.Normal(1, 0.1))
    amp1 = numpyro.sample('amp1', dist.Uniform(0, 1))
    amp2 = numpyro.sample('amp2', dist.Uniform(0, 1))
    wid1 = numpyro.sample('wid1', dist.Uniform(0.5, 10))
    wid2 = numpyro.sample('wid2', dist.Uniform(0.5, 10))

    # Model definition
    with numpyro.plate(f'epoch=1..{n_epoch}', n_epoch, dim=-2):
        mean1 = numpyro.sample('mean1', dist.Uniform(cen - 0.2 * delta_cen, cen + 0.2 * delta_cen))
        mean2 = numpyro.sample('mean2', dist.Uniform(cen - 0.2 * delta_cen, cen + 0.2 * delta_cen))
        with numpyro.plate(f'wavelength=1..{n_wavelength}', n_wavelength, dim=-1):
            comp1 = -amp1 * jnp.exp(dist.Normal(mean1, wid1).log_prob(wavelength))
            comp2 = -amp2 * jnp.exp(dist.Normal(mean2, wid2).log_prob(wavelength))
            pred = continuum + comp1 + comp2
            numpyro.deterministic('pred_1', continuum + comp1)
            numpyro.deterministic('pred_2', continuum + comp2)
            model = numpyro.deterministic('pred', pred)
            numpyro.sample('obs', dist.Normal(model, 0.05), obs=fluxes)

numpyro.render_model(sb2_model, model_args=(wavelengths, fluxes),
                 render_distributions=True, render_params=True)
```

```python
from jax import random
from numpyro.infer import MCMC, NUTS
rng_key = random.PRNGKey(0)

kernel = NUTS(sb2_model)
mcmc = MCMC(kernel,
            num_warmup=5000,
            num_samples=5000)
# print(type(wavelengths[0]))
mcmc.run(rng_key, wavelength=jnp.array(wavelengths), fluxes=jnp.array(fluxes))
mcmc.print_summary()
trace = mcmc.get_samples()
```

```python
def model(λ=None, fλ=None):
    nτ, nλ = λ.shape

    # spectral window
    λ0 = numpyro.param("λ0", 0.5 * (jnp.max(λ) + jnp.min(λ)))
    dλ0 = numpyro.param("dλ0", (jnp.max(λ) - jnp.min(λ)))

    # Continuum
    ε = numpyro.sample('ε', dist.Normal(1.0, 0.1))

    # components
    A1 = numpyro.sample('𝛼1', dist.Uniform(0.1, 1.0))
    A2 = numpyro.sample('𝛼2', dist.Uniform(0.1, 1.0))
    σ1 = numpyro.sample('σ1', dist.Uniform(0.5, 20))
    σ2 = numpyro.sample('σ2', dist.Uniform(0.5, 20))


    with numpyro.plate(f'τ=1..{nτ}', nτ, dim=-2):
        λ1_τ = numpyro.sample("λ1_τ", dist.Uniform(λ0 - 1./4. * dλ0, λ0 + 1./4. * dλ0))
        λ2_τ = numpyro.sample("λ2_τ", dist.Uniform(λ0 - 0.4 * dλ0, λ0 + 0.4 * dλ0))

        with numpyro.plate(f'λ=1..{nλ}', nλ, dim=-1):
            comp1 = numpyro.deterministic("c1",
                        -A1 * jnp.exp(dist.Normal(λ1_τ, σ1).log_prob(λ)))
            comp2 = numpyro.deterministic("c2",
                        -A2 * jnp.exp(dist.Normal(λ2_τ, σ2).log_prob(λ)))
            fλ_pred = numpyro.deterministic("fλ_pred", ε + comp1 + comp2)
            numpyro.sample("fλ", dist.Normal(fλ_pred, 0.05), obs=fλ)

numpyro.render_model(model, model_args=(wavelengths, fluxes),
                 render_distributions=False, render_params=True)
```

```python
def model1(λ=None, fλ=None, K=2):
    nτ, nλ = λ.shape

    # spectral window
    λ0 = numpyro.param("λ0", 0.5 * (jnp.max(λ) + jnp.min(λ)))
    dλ0 = numpyro.param("dλ0", (jnp.max(λ) - jnp.min(λ)))

    # Continuum
    ε = numpyro.sample('ε', dist.Normal(1.0, 0.1))

    with numpyro.plate(f"k=1..{K}", K, dim=-3):
        Ak = numpyro.sample('𝛼_k', dist.Uniform(0.0, 1.0))
        σk = numpyro.sample('σ_k', dist.Uniform(0.5, 20))

        with numpyro.plate(f'τ=1..{nτ}', nτ, dim=-2):
            λ_τk = numpyro.sample("λ_τk", dist.Uniform(λ0 - 1./4. * dλ0, λ0 + 1./4. * dλ0))

            with numpyro.plate(f'λ=1..{nλ}', nλ, dim=-1):
                Ck = numpyro.deterministic("C_λk",
                    -Ak * jnp.exp(dist.Normal(λ_τk, σk).log_prob(λ)))
                fλ_pred = numpyro.deterministic("fλ_pred", ε + Ck.sum(axis=-3))

                numpyro.sample("fλ", dist.Normal(fλ_pred, 0.05), obs=fλ)

numpyro.render_model(model1, model_args=(wavelengths, fluxes),
                 render_distributions=True, render_params=True)
```

```python

rng_key = random.PRNGKey(0)
kernel = NUTS(model1)
mcmc = MCMC(kernel, num_warmup=5000, num_samples=1000)
mcmc.run(rng_key, λ=wavelengths, fλ=fluxes)
mcmc.print_summary()
trace = mcmc.get_samples()
```
